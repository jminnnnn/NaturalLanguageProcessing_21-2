````python
# 파일을 불러들어와 json파일을 파싱하여 필요한 텍스트를 추출 & # 추출된 텍스트를 preprocessing
import json
import re
from jamo import h2j, j2hcj
with open('NIRW1900000011.json', 'rt', encoding='UTF8') as json_file:
    data = json.load(json_file)
    
words_list=[]
letter_set=[]
for i in range(len(data['document'])):
    each_document=data['document'][i]
    for j in range(len(each_document['paragraph'])):
#         print(len(each_document['paragraph']))
#         print(e)
        each_paragraph = each_document['paragraph'][j]
        for k in range(len(each_paragraph['form'].split(' '))):
                each_word=each_paragraph['form'].split(' ')[k]
                each_word_preprocessed=re.sub('[^ ㄱ-ㅣ 가-힣]','',each_word)
                if len(each_word_preprocessed) != 0:
                    if len(j2hcj(h2j(each_word_preprocessed))) == 1:
                        continue
                    words_list.append(each_word_preprocessed)
#                     print(type(each_word_preprocessed))
                    for l in range(len(each_word_preprocessed)):
#                         print(each_word_preprocessed)
                        each_letter_preprocessed = list(each_word_preprocessed)[l]
                        letter_set.append(each_letter_preprocessed)
#                         print(each_spelling_preprocessed)

print(len(words_list))
print(len(letter_set))
````

````python
# 정제된 데이터에 대한 전체 토큰수, 글자 수 등의 기본적인 통계작업 및 출력
print('한글 단어수는 {}개'.format(len(words_list)))
print('한글 글자수는 {}개'.format(len(letter_set)))
````

````python
#받침이 없는 글자를 찾아 빈도 순으로 출력
#받침 유무 확인
from collections import Counter

def has_no_coda(letter):
    return (ord(letter[-1])-44032) %28 ==0
def count_num(list):
    num_list={}
    count_list=Counter(list)
    for i in list:
        num_list[i]=count_list[i]
    return num_list
        
no_coda=[]
for i in range(len(letter_set)):
    if has_no_coda(letter_set[i])==True:
        no_coda.append(letter_set[i])
# count_num('박')
# print(count_num('박','흥'))
````

````python
letter_order = sorted(count_num(no_coda), key=count_num(no_coda).get,reverse=True)
print(letter_order)
````

````python
# 종성 리스트. 00 ~ 27 + 1(1개 없음)
JONGSUNG_LIST = [' ', 'ㄱ', 'ㄲ', 'ㄳ', 'ㄴ', 'ㄵ', 'ㄶ', 'ㄷ', 'ㄹ', 'ㄺ', 'ㄻ', 'ㄼ', 'ㄽ', 'ㄾ', 'ㄿ', 'ㅀ', 'ㅁ', 'ㅂ', 'ㅄ', 'ㅅ', 'ㅆ', 'ㅇ', 'ㅈ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']
# 겹받침 종성 리스트 . 00 ~ 12
TWIN_JONGSUNG_LIST = ['ㄲ', 'ㄳ', 'ㄵ', 'ㄶ', 'ㄺ', 'ㄻ', 'ㄼ', 'ㄽ', 'ㄾ',  'ㄿ', 'ㅀ', 'ㅄ', 'ㅆ']
````

````python
twin_jongsung_ord=[]
for i in range(len(JONGSUNG_LIST)):
    each_JONGSUNG=JONGSUNG_LIST[i]
    if each_JONGSUNG in TWIN_JONGSUNG_LIST:
        twin_jongsung_ord.append((i))
````

````python
def has_twin_coda(letter):
    return (ord(letter[-1])-44032) %28 in twin_jongsung_ord
def count_num(list):
    num_list={}
    count_list=Counter(list)
    for i in list:
        num_list[i]=count_list[i]
    return num_list
        
twin_coda=[]
for i in range(len(letter_set)):
    if has_twin_coda(letter_set[i])==True:
        twin_coda.append(letter_set[i])
````

````python
twin_letter_order = sorted(count_num(twin_coda), key=count_num(twin_coda).get,reverse=True)
print(twin_letter_order)
````

````python
def find_love(word):
    objective=j2hcj(h2j(word))
    return objective[0]=='ㅅ' and objective[-1]=='ㅇ'
def count_num(list):
    num_list={}
    count_list=Counter(list)
    for i in list:
        num_list[i]=count_list[i]
    return num_list
love_list=[]
for i in range(len(words_list)):
    if find_love(words_list[i])==True:
        love_list.append(words_list[i])
````

````python
love_order=sorted(count_num(love_list),key=count_num(love_list).get, reverse=True)
print(love_order)
````

````python
# 받침이 없는 글자로 이루어진 어절을 찾아 (예: 가자)그 빈도 순으로 출력

#단어를 글자로 나누고, 모든 글자가 받침이 없는지 확인하는 함수
def word_no_coda(word):
    for i in range(len(word)):
        if has_no_coda(word[i])==True:
            continue
        else:
            return False
    return True
def count_num(list):
    num_list={}
    count_list=Counter(list)
    for i in list:
        num_list[i]=count_list[i]
    return num_list

no_coda_list=[]
for i in range(len(words_list)):
    if word_no_coda(words_list[i])==True:
        no_coda_list.append(words_list[i])
````

````python
no_coda_order=sorted(count_num(no_coda_list),key=count_num(no_coda_list).get, reverse=True)
print(no_coda_order)
````
